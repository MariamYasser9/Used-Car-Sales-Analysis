# Visualization

```python
sns.set(style="whitegrid")
plt.style.use("seaborn-v0_8-whitegrid")

palette_brown = sns.color_palette(["#4E342E", "#795548", "#A1887F", "#D7CCC8", "#EFEBE9"])
```

```python
# Create Car Age from Manufactured Year
current_year = 2024
df["Car Age"] = current_year - df["Manufactured Year"]

# Remove invalid or zero prices/mileage
df = df[(df["Price-$"] > 0) & (df["Mileage-KM"] > 0)]
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Price-$"], bins=40, kde=True, color="#6D4C41")
plt.title("Distribution of Car Prices", fontsize=14, weight='bold')
plt.xlabel("Price-$")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Purchased Price-$"], bins=40, kde=True, color="#795548")
plt.title("Distribution of Purchased Price-$", fontsize=14, weight='bold')
plt.xlabel("Purchased Price ($)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=df, y="Purchased Price-$", color="#6D4C41")
plt.title("Outliers in Purchased Price-$", fontsize=14, weight='bold')
plt.ylabel("Purchased Price ($)")
plt.show()
```

```python
Q1 = df["Purchased Price-$"].quantile(0.01)
Q3 = df["Purchased Price-$"].quantile(0.99)
df_filtered = df[(df["Purchased Price-$"] >= Q1) & (df["Purchased Price-$"] <= Q3)]

plt.figure(figsize=(8,5))
sns.histplot(df_filtered["Purchased Price-$"], bins=40, kde=True, color="#A1887F")
plt.title("Purchased Price-$ (After Removing Extreme Outliers)", fontsize=14, weight='bold')
plt.xlabel("Purchased Price ($)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="Car Age", y="Price-$", alpha=0.7, color="#8D6E63")
sns.regplot(data=df, x="Car Age", y="Price-$", scatter=False, color="#4E342E")
plt.title("Price vs. Car Age", fontsize=14, weight='bold')
plt.xlabel("Car Age (Years)")
plt.ylabel("Price ($)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="Mileage-KM", y="Price-$", alpha=0.7, color="#A1887F")
sns.regplot(data=df, x="Mileage-KM", y="Price-$", scatter=False, color="#4E342E")
plt.title("Price vs. Mileage (KM)", fontsize=14, weight='bold')
plt.xlabel("Mileage (KM)")
plt.ylabel("Price ($)")
plt.show()
```

```python
top_manufacturers = df["Manufacturer Name"].value_counts().nlargest(10).index

manufacturer_median = (
    df[df["Manufacturer Name"].isin(top_manufacturers)]
    .groupby("Manufacturer Name")["Price-$"]
    .median()
    .sort_values(ascending=False)
    .reset_index()
)

plt.figure(figsize=(10,5))
sns.barplot(
    data=manufacturer_median,
    x="Manufacturer Name",
    y="Price-$",
    palette=palette_brown
)
plt.title("Median Price by Top 10 Manufacturers", fontsize=14, weight='bold')
plt.xlabel("Manufacturer")
plt.ylabel("Median Price ($)")
plt.xticks(rotation=45)
plt.show()
```

```python
top_manufacturers = df["Manufacturer Name"].value_counts().nlargest(10).index
plt.figure(figsize=(10,5))
sns.barplot(
    data=df[df["Manufacturer Name"].isin(top_manufacturers)],
    x="Manufacturer Name",
    y="Price-$",
    estimator=np.median,
    palette=palette_brown
)
plt.title("Median Price by Top 10 Manufacturers", fontsize=14, weight='bold')
plt.xlabel("Manufacturer")
plt.ylabel("Median Price ($)")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="Car Type", y="Price-$", palette=palette_brown)
plt.title("Price Distribution by Car Type", fontsize=14, weight='bold')
plt.xlabel("Car Type")
plt.ylabel("Price ($)")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(data=df, x="Gearbox", y="Price-$", palette=palette_brown)
plt.title("Price by Gearbox Type", fontsize=14, weight='bold')
plt.xlabel("Gearbox")
plt.ylabel("Price ($)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="Car Age", y="Mileage-KM", alpha=0.7, color="#8D6E63")
sns.regplot(data=df, x="Car Age", y="Mileage-KM", scatter=False, color="#4E342E")
plt.title("Car Age vs. Mileage", fontsize=14, weight='bold')
plt.xlabel("Car Age (Years)")
plt.ylabel("Mileage (KM)")
plt.show()
```

```python
plt.figure(figsize=(9,7))
corr = df[["Price-$","Purchased Price-$","Mileage-KM","Engine Power-HP","Car Age","Margin-%","Sales Rating","Sales Commission-$"]].corr()
sns.heatmap(corr, annot=True, cmap="BrBG", fmt=".2f")
plt.title("Correlation Matrix", fontsize=14, weight='bold')
plt.show()
```

```python
sale_counts = df["Car Sale Status"].value_counts()
plt.figure(figsize=(6,6))
plt.pie(
      sale_counts,
      labels=sale_counts.index,
      autopct='%1.1f%%',
      startangle=90,
      colors=["#795548", "#A1887F"],
      textprops={'fontsize': 12}
    )
plt.title("Distribution of Sold vs Unsold Cars", fontsize=14, weight='bold')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Margin-%"], bins=40, kde=True, color="#795548")
plt.title("Distribution of Profit Margin (%)", fontsize=14, weight='bold')
plt.xlabel("Margin (%)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="Sales Rating", y="Price-$", alpha=0.7, color="#A1887F")
sns.regplot(data=df, x="Sales Rating", y="Price-$", scatter=False, color="#4E342E")
plt.title("Sales Rating vs Price", fontsize=14, weight='bold')
plt.xlabel("Sales Rating")
plt.ylabel("Price ($)")
plt.show()
```

```python
top_agents = df["Sales Agent Name"].value_counts().nlargest(10).index
df_agents = (
    df[df["Sales Agent Name"].isin(top_agents)]
    .groupby("Sales Agent Name")["Margin-%"]
    .mean()
    .sort_values(ascending=False)
    .reset_index()
)

plt.figure(figsize=(10,5))
sns.barplot(
    data=df_agents,
    x="Sales Agent Name",
    y="Margin-%",
    palette=palette_brown
)
plt.title("Average Margin (%) by Top 10 Sales Agents", fontsize=14, weight='bold')
plt.xlabel("Sales Agent Name")
plt.ylabel("Average Margin (%)")
plt.xticks(rotation=45)
plt.show()
```

```python
df_valid_years = df[df["Sold Year"] > 1970]

plt.figure(figsize=(10,5))
sns.lineplot(
    data=df_valid_years,
    x="Sold Year",
    y="Price-$",
    estimator="mean",
    color="#4E342E",
    marker="o"
)
plt.title("Average Car Price Over the Years (Valid Sales Only)", fontsize=14, weight='bold')
plt.xlabel("Sold Year")
plt.ylabel("Average Price ($)")
plt.show()
```

```python
df['Year'] = pd.to_datetime(df['Sold Date'], errors='coerce').dt.year

sales_per_year = (
    df[(df['Car Sale Status'].str.lower() == 'sold') & (df['Year'] > 1970)]
    .groupby('Year')
    .size()
)

plt.figure(figsize=(10, 6))
plt.plot(sales_per_year.index, sales_per_year.values, marker='o', color="#6D4C41", linewidth=2)

plt.title('Total Car Sales Per Year', fontsize=16, weight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Cars Sold', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
sold_cars = df[df["Car Sale Status"].str.lower() == "sold"]

top_models = sold_cars["Car Name"].value_counts().nlargest(10).reset_index()
top_models.columns = ["Car Name", "Count"]

plt.figure(figsize=(10,5))
sns.barplot(
    data=top_models,
    x="Car Name",
    y="Count",
    palette=palette_brown
)
plt.title("Top 10 Sold Car Models", fontsize=14, weight='bold')
plt.xlabel("Car Model")
plt.ylabel("Number of Cars Sold")
plt.xticks(rotation=45)
plt.show()
```
